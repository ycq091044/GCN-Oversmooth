```python
def get(Dataset, layers):
    loss_train, acc_train, loss_val, acc_val, Fwise, Nwise = [], [], [], [], [], []
    for file in range(20):
        tloss_train, tacc_train, tloss_val, tacc_val, tFwise, tNwise = [], [], [], [], [], []
        with open(Dataset + '/output_{}_{}.log'.format(file, layers), 'r') as infile:
            for i in infile:
                try:
                    temp = i.split('-')
                    if len(temp) == 6:
                        a, b, c, d, e, f = list(map(float, temp))
                    elif len(temp) == 7:
                        temp[0] = temp[0] + '-' + temp[1]
                        a, exp, b, c, d, e, f = list(map(float, temp))
                    tloss_train.append(a)
                    tacc_train.append(b)
                    tloss_val.append(c)
                    tacc_val.append(d)
                    tFwise.append(e)
                    tNwise.append(f)
                except:
                    continue
        loss_train.append(tloss_train)
        acc_train.append(tacc_train)
        loss_val.append(tloss_val)
        acc_val.append(tacc_val)
        Fwise.append(tFwise)
        Nwise.append(tNwise)
        
    return loss_train, acc_train, loss_val, acc_val, Fwise, Nwise
```

```python
Dataset = 'cora' # 'citeseer', 'pubmed'
```

```python
import matplotlib.pyplot as plt
import numpy as np
    
ceiling = 3

plt.figure(figsize=(20, 3.8))
plt.rc('font', size=15)
from matplotlib import rcParams

rcParams['font.family'] = 'sans-serif'


plt.subplot(151)
loss_train, acc_train, loss_val, acc_val, Fwise, Nwise = get(Dataset, 0)
loss_train = np.array(loss_train); loss_val = np.array(loss_val)
lossTGCN_mean = loss_train.mean(axis=0)
lossTGCN_std = loss_train.std(axis=0)
lossVGCN_mean = loss_val.mean(axis=0)
lossVGCN_std = loss_val.std(axis=0)

plt.plot(lossTGCN_mean)
plt.fill_between(np.arange(len(lossTGCN_mean)), lossTGCN_mean - lossTGCN_std, lossTGCN_mean + lossTGCN_std, alpha = 0.3)
plt.plot(lossVGCN_mean)
plt.fill_between(np.arange(len(lossVGCN_mean)), lossVGCN_mean - lossVGCN_std, lossVGCN_mean + lossVGCN_std, alpha = 0.3)
plt.ylim([0, ceiling])

plt.legend(['Training Loss', 'Test Loss'], frameon=False)
plt.ylabel('Loss Function')
plt.xlabel('Epochs ({})'.format(Dataset))
plt.title('2 layers')

plt.xticks(np.array([0, 250, 500, 750, 1000]), np.array([0, 250, 500, 750, 1000]))


plt.subplot(152)
loss_train, acc_train, loss_val, acc_val, Fwise, Nwise = get(Dataset, 1)
loss_train = np.array(loss_train); loss_val = np.array(loss_val)
lossTGCN_mean = loss_train.mean(axis=0)
lossTGCN_std = loss_train.std(axis=0)
lossVGCN_mean = loss_val.mean(axis=0)
lossVGCN_std = loss_val.std(axis=0)

plt.plot(lossTGCN_mean)
plt.fill_between(np.arange(len(lossTGCN_mean)), lossTGCN_mean - lossTGCN_std, lossTGCN_mean + lossTGCN_std, alpha = 0.3)
plt.plot(lossVGCN_mean)
plt.fill_between(np.arange(len(lossVGCN_mean)), lossVGCN_mean - lossVGCN_std, lossVGCN_mean + lossVGCN_std, alpha = 0.3)
plt.ylim([0, ceiling])

plt.legend(['Training Loss', 'Test Loss'], frameon=False)
plt.xlabel('Epochs ({})'.format(Dataset))
plt.xticks(np.array([0, 250, 500, 750, 1000]), np.array([0, 250, 500, 750, 1000]))
plt.title('3 layers')

plt.subplot(153)
loss_train, acc_train, loss_val, acc_val, Fwise, Nwise = get(Dataset, 3)
loss_train = np.array(loss_train); loss_val = np.array(loss_val)
lossTGCN_mean = loss_train.mean(axis=0)
lossTGCN_std = loss_train.std(axis=0)
lossVGCN_mean = loss_val.mean(axis=0)
lossVGCN_std = loss_val.std(axis=0)

plt.plot(lossTGCN_mean)
plt.fill_between(np.arange(len(lossTGCN_mean)), lossTGCN_mean - lossTGCN_std, lossTGCN_mean + lossTGCN_std, alpha = 0.3)
plt.plot(lossVGCN_mean)
plt.fill_between(np.arange(len(lossVGCN_mean)), lossVGCN_mean - lossVGCN_std, lossVGCN_mean + lossVGCN_std, alpha = 0.3)
plt.ylim([0, ceiling])
plt.legend(['Training Loss', 'Test Loss'], frameon=False)
plt.xlabel('Epochs ({})'.format(Dataset))
plt.xticks(np.array([0, 250, 500, 750, 1000]), np.array([0, 250, 500, 750, 1000]))
plt.title('5 layers')

plt.subplot(154)
loss_train, acc_train, loss_val, acc_val, Fwise, Nwise = get(Dataset, 8)
loss_train = np.array(loss_train); loss_val = np.array(loss_val)
lossTGCN_mean = loss_train.mean(axis=0)
lossTGCN_std = loss_train.std(axis=0)
lossVGCN_mean = loss_val.mean(axis=0)
lossVGCN_std = loss_val.std(axis=0)

plt.plot(lossTGCN_mean)
plt.fill_between(np.arange(len(lossTGCN_mean)), lossTGCN_mean - lossTGCN_std, lossTGCN_mean + lossTGCN_std, alpha = 0.3)
plt.plot(lossVGCN_mean)
plt.fill_between(np.arange(len(lossVGCN_mean)), lossVGCN_mean - lossVGCN_std, lossVGCN_mean + lossVGCN_std, alpha = 0.3)
plt.ylim([0, ceiling])

plt.legend(['Training Loss', 'Test Loss'], frameon=False)
plt.xlabel('Epochs ({})'.format(Dataset))
plt.xticks(np.array([0, 250, 500, 750, 1000]), np.array([0, 250, 500, 750, 1000]))
plt.title('10 layers')

plt.subplot(155)
loss_train, acc_train, loss_val, acc_val, Fwise, Nwise = get(Dataset, 48)
loss_train = np.array(loss_train); loss_val = np.array(loss_val)
lossTGCN_mean = loss_train.mean(axis=0)
lossTGCN_std = loss_train.std(axis=0)
lossVGCN_mean = loss_val.mean(axis=0)
lossVGCN_std = loss_val.std(axis=0)

plt.plot(lossTGCN_mean)
plt.fill_between(np.arange(len(lossTGCN_mean)), lossTGCN_mean - lossTGCN_std, lossTGCN_mean + lossTGCN_std, alpha = 0.3)
plt.plot(lossVGCN_mean)
plt.fill_between(np.arange(len(lossVGCN_mean)), lossVGCN_mean - lossVGCN_std, lossVGCN_mean + lossVGCN_std, alpha = 0.3)
plt.ylim([0, ceiling])
plt.legend(['Training Loss', 'Test Loss'], frameon=False)
plt.xlabel('Epochs ({})'.format(Dataset))
plt.title('50 layers')

plt.tight_layout()
plt.xticks(np.array([0, 250, 500, 750, 1000]), np.array([0, 250, 500, 750, 1000]))

plt.savefig('loss-{}.pdf'.format(Dataset))
```
